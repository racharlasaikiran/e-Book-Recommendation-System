```python
df['average_rating']=df['average_rating'].astype(float)
```

```python
df.loc[[3348,4702,5877,8979 ]]
```

```python
df.drop(columns=['Unnamed: 12'],inplace=True)
```

```python
df['num_pages']=df['num_pages'].astype(int)
```

```python
df['authors'] = df['authors'].str.replace('/', ', ')
```

```python
df['publisher'] = df['publisher'].str.replace('.', '', regex=False)
```

```python
df['publication_year'] = pd.to_datetime(df['publication_date'], errors='coerce').dt.year
```

```python
df['publication_year'] = pd.to_numeric(df['publication_year'], errors='coerce').astype('Int64')
```

```python
df['average_rating_cat']=pd.cut(df['average_rating'],bins=[-1,1,2,3,4,5],
                       labels=['worst','bad','average','good','excellent'])
```

```python
df['ratings_count_cat']=pd.cut(df['ratings_count'],bins=[-1,1000000,2000000,3000000,4000000,5000000],
                       labels=['very low','low','normal','high','very high'])
```

```python
df['page_category'] = pd.cut(df['num_pages'], bins=[-1, 100, 500, 1000, 7000],
                             labels=['Short', 'Medium', 'Long', 'Very Long'])
```

```python
df["author_publisher_year"] = (
    df["authors"] + " , " +
    df["publisher"] + ", " +
    df['publication_year'].astype(str) + ", " +
    df['ratings_count_cat'].astype(str) + ", " +
    df['page_category'].astype(str) + ", " +
    df['average_rating_cat'].astype(str) + ", " +
    df['language_code']
)
```

```python
df['author_publisher_year'].head()
```

```python
df=df.drop(columns=['publication_year','average_rating_cat','ratings_count_cat','page_category'])
```

# M

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vec = TfidfVectorizer()
feature_vectors = vec.fit_transform(df['author_publisher_year'])
print(feature_vectors[0])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(feature_vectors)
print(similarity[0])
```

```python
print(similarity.shape)
```

```python
list_of_books = df['title'].tolist()
print(list_of_books[0:10])
```

```python
book_name = input('Enter any book name: ')
```

```python
import difflib
close_match = difflib.get_close_matches(book_name,list_of_books)
print(close_match)
```

```python
close_one = close_match[0]
print(close_one)
```

```python
index_of_book = df[df.title == close_one]['bookID'].index[0]
print(index_of_book)
```

```python
similarity_score = list(enumerate(similarity[index_of_book]))
print(similarity_score[0:10])
```

```python
len(similarity_score)
```

```python
sorted_similar_books = sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_books[0:10])
```

```python
li =[]
for i in sorted_similar_books:
    li.append(i[0])
    if len(li)==10:
        break
```
